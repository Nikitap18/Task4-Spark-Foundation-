```python
# TO Check missing values
missing_values=(((terr_df.isnull().sum().sum())/terr_df.size)*100)
missing_values
```

### Inference: 
* More than 50% values are Null. we need to clean the dataset. 

```python
# creating dataframe with necessary columns only.
terr_df=terr_df[['Year','Month','Day','Country','State','Region','Attack_Type','Target','Killed','Wounded','Group','Target_Type','Weapon_Type','Latitude','Longitude','City']]
terr_df.head(10)

```

```python
for i in terr_df.columns:
    print(i,terr_df[i].nunique())
```

```python
terr_df.info()
```

```python
terr_df['Wounded']=terr_df['Wounded'].fillna(0).astype(int)
terr_df['Killed']=terr_df['Killed'].fillna(0).astype(int)
```

```python
terr_df.info()
```

```python
# cleaned data
terr_df.head(10)
```

```python
x_year=terr_df['Year'].unique()
y_count_years=terr_df['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(12,3))
sns.barplot(x=x_year,
           y=y_count_years,
           palette='dark:salmon_r')
plt.title('Attack_of_Years',fontsize=15)
plt.xlabel('Attack of Year')
plt.ylabel('No. of Attack each Year',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

#### This Graph Shows
* 2014 becomes witness of a huge terrorist attacks.
* 2014 onwards recorded a large increase in attacks of terrorism

```python
plt.figure(figsize=(12,3))
terr_df.groupby(['Year'])['Killed'].sum().plot(kind='bar',colormap='summer')

plt.title('No. of deaths in different years',fontsize=15)
plt.xlabel('Year')
plt.ylabel('No. of Deaths',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

#### Inference:
* Number of deaths caused by terrorism is more between 2014 and 2018 with on average >50k people being killed.
High peak went from 2014 with 45K+ deaths.

```python
plt.figure(figsize=(15,5))
sns.kdeplot(terr_df['Year'],hue=terr_df['Region'])
plt.title('Terrorist acticities by region in each year',fontsize=15)
plt.xlabel('Years')
plt.ylabel('Frequency of Attacks')
plt.xticks(rotation=90)
plt.show()
```

# Bivariate Analysis

```python
plt.subplot(1,2,1)
terr_df.groupby(['Region'])['Wounded'].sum().sort_values(ascending=False).plot(kind='bar',color='pink')
plt.title('Region having Wounded people',fontsize=15)
plt.xlabel('Region')
plt.ylabel('Number of Wounded',fontsize=15)
plt.xticks(rotation=90)

plt.subplot(1,2,2)
terr_df.groupby(['Region'])['Killed'].sum().sort_values(ascending=False).plot(kind='bar',colormap='summer')
plt.title('Region having Killed  people',fontsize=15)
plt.xlabel('Region')
plt.ylabel('Number of Killed',fontsize=15)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15, 5)
```

### Inference:
* Middle East & North Africa has the most killed people (>1.4 lacs) and wounded people(>2 lacks)

```python
plt.subplot(1,2,2)
terr_df.groupby(['Country'])['Wounded'].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='pink')
plt.title('Country having Wounded people',fontsize=15)
plt.xlabel('Country')
plt.ylabel('Number of Wounded',fontsize=15)
plt.xticks(rotation=90)

plt.subplot(1,2,1)
terr_df.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='y')
plt.title('Country having Killed  people',fontsize=15)
plt.xlabel('Country')
plt.ylabel('Number of Killed',fontsize=15)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15, 5)
```

### Inference:
* Iraq has the most killed people (>70K) and wounded people (>1.2 lacks). Afghanistan,Pakistan and India all saw peak death while Iraq has overtaken all of these at top of the table, suffered 37,700 more fatalities

```python
plt.figure(figsize=(5,3))
sns.barplot(terr_df['Group'].value_counts()[1:11].values,terr_df['Group'].value_counts()[1:11].values,palette='viridis')
plt.title('Top 10 Terrorist Organization With Highest Terror Attacks')
plt.xlabel('No of Attacks')
plt.ylabel('Terrorist Group')
plt.xticks(rotation=90)
plt.show()
```

```python
terr_df_tal=terr_df[terr_df.Group=='Taliban']
```

```python
plt.figure(figsize=(5,3))
terr_df_tal.groupby(['Year'])['Killed'].sum().plot(kind='bar',color='Violet')
plt.title('People Killed by Taliban over the year')
plt.xlabel('Year')
plt.ylabel('No of people killed')
plt.xticks(rotation=90)
plt.show()
```

### Inference:

```python
plt.subplot(1,2,2)
terr_df_2014.groupby(['Country'])['Wounded'].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='pink')
plt.title('Top 10 Countries with max Wounded people',fontsize=15)
plt.xlabel('Country')
plt.ylabel('Number of people Wounded',fontsize=15)
plt.xticks(rotation=90)

plt.subplot(1,2,1)
terr_df_2014.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='y')
plt.title('Top 10 Countries with max killed people',fontsize=15)
plt.xlabel('Country')
plt.ylabel('Number of  people Killed',fontsize=15)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15, 5)
```
